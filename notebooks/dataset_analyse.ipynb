```python
import glob
import cv2
import os
import json
```

```python
def create_img_list(folder_list, output_folder_path, class_names_to_indices, split_name):
    image_paths_to_class_names = {}
    for folder_path in folder_list:
        # folder_path = os.path.join(folder_list, folder)
        class_name = os.path.basename(folder_path)
        image_files = glob.glob(os.path.join(folder_path, "*.jpg"))
        for image_file in image_files:
            image_paths_to_class_names[image_file] = class_names_to_indices[class_name]
    with open(os.path.join(output_folder_path, f'{split_name}_list.json'), 'w') as f:
        json.dump(image_paths_to_class_names, f)


def getfolder_path_list(folder_path):
    folder_list = sorted(os.listdir(folder_path) )
    folder_list = [os.path.join(folder_path, folder) for folder in folder_list if os.path.isdir(os.path.join(folder_path, folder))]
    return folder_list

def create_dataset_json(dataset_folder_path, train_folder_name, test_folder_name, validation_folder_name, output_folder_path):
    
    train_folder_path = os.path.join(dataset_folder_path, train_folder_name)
    test_folder_path = os.path.join(dataset_folder_path, test_folder_name)
    validation_folder_path = os.path.join(dataset_folder_path, validation_folder_name)
    
    class_names = os.listdir(train_folder_path)
    
    train_folder_list = getfolder_path_list(train_folder_path)      
    test_folder_list = getfolder_path_list(test_folder_path)
    validation_folder_list = getfolder_path_list(validation_folder_path)
    
    # first save the class_names_to_indices to a json file
    class_names_to_indices = {}
    for idx, class_name in enumerate(class_names):
        class_names_to_indices[class_name] = idx
    with open(os.path.join(output_folder_path, 'class_names_to_indices.json'), 'w') as f:
        json.dump(class_names_to_indices, f)
        
    create_img_list(train_folder_list, output_folder_path, class_names_to_indices, 'train')
    create_img_list(test_folder_list, output_folder_path, class_names_to_indices, 'test')
    create_img_list(validation_folder_list, output_folder_path, class_names_to_indices, 'val')
    
dataset_folder_path = "/hdd/side_projects/data/fruits_360/fruits-360_dataset_original-size/fruits-360-original-size"
train_folder_name = "Training"
test_folder_name = "Test"
validation_folder_name = "Validation"

output_folder_path = "/hdd/side_projects/mmlab_tutorials/data"

create_dataset_json(dataset_folder_path, train_folder_name, test_folder_name, validation_folder_name, output_folder_path)
```
